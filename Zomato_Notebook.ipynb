```python

# 1. List all restaurants in a specific locality.
locality = "SM Megamall| Ortigas| Mandaluyong City"

result = df[df["Locality"] == locality][["RestaurantName","Rating"]]

print(result)

```

```python
# 2. Show the name and city of restaurants with a rating lower than 3.

df[df['Rating']<3][['RestaurantName','City']]
```

```python
# 3. Count the number of restaurants that do not offer table booking.
a = df[df["Has_Table_booking"]=='No']["RestaurantID"].count()
print(a)
```

```python
# 4. Display restaurants that have an average cost for two less than $50.
currency_map = {
    "Botswana Pula(P)": "BWP",
    "Brazilian Real(R$)": "BRL",
    "Dollar($)": "USD",
    "Emirati Diram(AED)": "AED",
    "Indian Rupees(Rs.)": "INR",
    "Indonesian Rupiah(IDR)": "IDR",
    "NewZealand($)": "NZD",
    "Pounds(?)": "GBP",
    "Qatari Rial(QR)": "QAR",
    "Rand(R)": "ZAR",
    "Sri Lankan Rupee(LKR)": "LKR",
    "Turkish Lira(TL)": "TRY"
}
df["Currency_Code"] = df["Currency"].map(currency_map)
exchange_to_usd = {
    "USD": 1.00,
    "BWP": 0.079,
    "BRL": 0.19,
    "AED": 0.27,
    "INR": 0.012,
    "IDR": 0.000067,
    "NZD": 0.60,
    "GBP": 1.26,
    "QAR": 0.27,
    "ZAR": 0.054,
    "LKR": 0.0031,
    "TRY": 0.052
}
df["Cost_in_USD"] = df["Average_Cost_for_two"] * df["Currency_Code"].map(exchange_to_usd)
df[df["Cost_in_USD"]<50]["RestaurantName"]
```

```python
# 9. Display the top 10 restaurants by rating in a specific country.
CountryCode = "1"

result = (
    df[df["CountryCode"] == CountryCode]
    [["RestaurantName", "Rating"]]
    .sort_values(by="Rating", ascending=False)
    .head(10)
)

print(result)
```
